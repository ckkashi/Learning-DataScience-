```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ipl_dataset = pd.read_csv('ipl data.csv')
```

```python
ipl_dataset.head()
```

```python
ipl_dataset.shape
```

```python
match_cities = ipl_dataset['city'].value_counts()[:10]
```

```python
plt.figure(figsize=(20,10))
plt.bar(match_cities.keys(),match_cities.values,color='g')
plt.show()
```

```python
bat_first=ipl_dataset[ipl_dataset['win_by_runs']!=0]
```

```python
winners_by_bat_first = bat_first['winner'].value_counts()
winners_by_bat_first
```

```python
plt.figure(figsize=(15,10))
plt.bar(winners_by_bat_first.keys()[:5],winners_by_bat_first.values[:5],color='r')
plt.show()
```

```python
# for visualizing numerical values we use histogram
```

```python
plt.figure(figsize=(10,5))
plt.title('Win by Runs')
plt.xlabel('Runs')
plt.hist(bat_first['win_by_runs'],bins=30,color='red')
plt.show()
```

```python
plt.figure(figsize=(15,8))
winners_data = bat_first['winner'].value_counts()[:5]
# winners_data
plt.bar(winners_data.keys(),winners_data.values,color=['red','green','yellow','red','green'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
winners_data = bat_first['winner'].value_counts()[:5]
# winners_data
plt.pie(winners_data.values,labels=winners_data.keys(),autopct="%0.1f%%")
plt.show()
```

```python
ball_first = ipl_dataset[ ipl_dataset['win_by_wickets'] != 0 ]
```

```python
winners_by_ball_first = ball_first['winner'].value_counts()
```

```python
plt.figure(figsize=(15,5))
plt.bar(winners_by_ball_first.keys()[:5],winners_by_ball_first.values[:5],color='r')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(ball_first['win_by_wickets'],bins=50)
plt.title('Matches : Win by wickets ')
plt.xlabel('Wickets')
plt.ylabel('Matches')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.pie(winners_by_ball_first.values,labels=winners_by_ball_first.keys(),autopct="%0.1f%%")
plt.show()
```
