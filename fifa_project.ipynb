```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa_dataset = pd.read_csv('fifa data.csv')
```

```python
fifa_dataset.head()
```

```python
fifa_dataset.shape
```

```python
players_nationality = fifa_dataset['nationality'].value_counts()[:5]
```

```python
plt.bar(players_nationality.keys(),players_nationality,color='g')
plt.grid()
plt.show()
```

```python
players_salary = fifa_dataset[['short_name','wage_eur']][:10]
players_salary.sort_values(by='wage_eur',ascending=False) 
```
